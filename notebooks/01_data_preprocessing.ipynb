```python
from pathlib import Path
import sys

ROOT_DIR = Path.cwd().resolve()
while not (ROOT_DIR / "ml_core").exists() and ROOT_DIR != ROOT_DIR.parent:
    ROOT_DIR = ROOT_DIR.parent

if str(ROOT_DIR) not in sys.path:
    sys.path.insert(0, str(ROOT_DIR))

from ml_core.src.preprocessing.video_preprocessor import VideoPreprocessor
from ml_core.src.visualization.video_visualizer import VideoVisualizer
```

```python
processor = VideoPreprocessor(target_frames=16, target_size=(224, 224))
visualizer = VideoVisualizer()
sample_path = ROOT_DIR / "data" / "sample_videos" / "sample.mp4"

if sample_path.exists():
    processed_data = processor.preprocess_video(str(sample_path))
    if processed_data is not None:
        print(f"Tensor shape: {processed_data.shape}")
else:
    processed_data = None
    print(f"Sample video not found at: {sample_path}")
```

```python
if processed_data is not None:
    visualizer.plot_pixel_distribution(processed_data)
```

```python
if processed_data is not None:
    visualizer.plot_motion_energy(processed_data)
```
