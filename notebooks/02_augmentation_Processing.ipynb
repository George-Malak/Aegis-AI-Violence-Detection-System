```python
from pathlib import Path
import sys
from ml_core.src.data.dataset_loader import DatasetLoader
from ml_core.src.augmentation.video_augmentation import VideoAugmentor

ROOT_DIR = Path('/content/drive/MyDrive/DEPI Project').resolve()

Aegis_AI_VD_System_Root = Path('/content/Aegis-AI-Violence-Detection-System').resolve()
if str(Aegis_AI_VD_System_Root) not in sys.path:
    sys.path.insert(0, str(Aegis_AI_VD_System_Root))

if str(ROOT_DIR) not in sys.path:
    sys.path.insert(0, str(ROOT_DIR))
```

```python
augmentor = VideoAugmentor()

OUTP_dir = ROOT_DIR / "Train"/"Temp"
OUTP_dir.mkdir(parents=True, exist_ok=True)

loader = DatasetLoader(drive_root=[], class_names=["Abuse"])

dataset_path = [
    ROOT_DIR / "Train"
]

print(f"Initial loader.drive_root: {loader.drive_root}")
loader.drive_root = dataset_path
print(f"Modified loader.drive_root: {loader.drive_root}")
```

```python
try:
    dataset_manifest = loader.load_dataset()
    if dataset_manifest:
        for Video in dataset_manifest:
            video_path = Video["video_path"]
            original_video_name = Path(video_path).stem
            print(f"Original video name: {original_video_name}")
            class_name = Video["class_name"]
            output_directory = OUTP_dir / class_name

            augmentor.augment_video_pipeline(
                video_path=video_path,
                output_dir=str(output_directory),
                class_name=class_name,
                video_id=f"{original_video_name}",
                window_duration_sec=20,
                overlap_sec=0,
            )
        print(f"Video augmentation completed successfully for {len(dataset_manifest)} files!")
    else:
        print(f"No video files found under: {loader.drive_root}")
except FileNotFoundError as exc:
    print(str(exc))
```

```python
import os
from pathlib import Path # Ensure Path is imported for path operations

video = 0
# Iterate over immediate children names of ROOT_DIR (e.g., 'Train')
for classe_name in os.listdir(ROOT_DIR):
    full_path_to_classe = ROOT_DIR / classe_name
    print(f"Processing {full_path_to_classe} !!")
    print('-'*50)
    if full_path_to_classe.is_dir(): # Check if it's a directory (e.g., 'Train')
        # Iterate over immediate children names of the first-level directory (e.g., 'Abuse' within 'Train')
        for item_name in os.listdir(full_path_to_classe):
            full_path_to_item = full_path_to_classe / item_name

            if full_path_to_item.is_dir(): # Check if it's a directory (e.g., 'Abuse')
                # List the contents of the video category directory (i.e., the video files)
                cont = os.listdir(full_path_to_item)
                num_videos_in_category = len(cont)

                video += num_videos_in_category # Accumulate total video count
                print(f"Size of {item_name} is {num_videos_in_category} Video !")

print(f"Total number of videos found across all categories: {video}")
```
